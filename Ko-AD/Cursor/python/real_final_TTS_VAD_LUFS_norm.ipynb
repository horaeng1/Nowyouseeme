```python
import os
import json
from pathlib import Path

import torch
import torchaudio
import pandas as pd
import numpy as np
from pydub import AudioSegment
import ffmpeg
import pyloudnorm as pyln  # LUFS Ï∏°Ï†ï ÎùºÏù¥Î∏åÎü¨Î¶¨

from TTS.api import TTS  # Coqui TTS
```

```python
# =======================================================
# 0. Í≤ΩÎ°ú / Í∏∞Î≥∏ ÏÑ§Ï†ï
# =======================================================

JSON_PATH = "/content/drive/MyDrive/test_scene/description/gemini_BB_eng.json"  # AD JSON ÌååÏùº Í≤ΩÎ°ú
audio_path = "/content/drive/MyDrive/test_scene/Walter White Kills Mike Ehrmantraut.wav"  # ÏõêÎ≥∏ Ïò§ÎîîÏò§
input_video = "/content/drive/MyDrive/test_scene/Walter White Kills Mike Ehrmantraut.mp4"  # ÏõêÎ≥∏ ÎπÑÎîîÏò§

TTS_DIR = Path("/content/drive/MyDrive/test_scene/tts_from_json_vad_lufs_norm")
TTS_DIR.mkdir(exist_ok=True, parents=True)  # TTS ÌååÏùº Ï†ÄÏû• Ìè¥Îçî

final_audio_path = "/content/drive/MyDrive/test_scene/final_eng_female_BB.wav"
output_video = "/content/drive/MyDrive/test_scene/final_eng_female_BB.mp4"

# --- üîß AD ÌïÑÌÑ∞ÎßÅ / Î≥ºÎ•® ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ---
MIN_DESC_LEN = 2.5               # AD ÌÖçÏä§Ìä∏Ïùò ÏµúÏÜå ÏßÄÏÜç ÏãúÍ∞Ñ(Ï¥à)
DIALOGUE_LUFS_MARGIN = 2.0        # ADÎ•º Î∞∞Í≤Ω Î∞úÌôîÎ≥¥Îã§ Î™á LUFS ÏûëÍ≤å Ìï†ÏßÄ (2~4 LUFS Í∂åÏû•)
AD_OVERLAP_EXTRA_ATTENUATION_DB = -3.0  # üéØ ÎåÄÏÇ¨ÏôÄ Í≤πÏπòÎäî Íµ¨Í∞ÑÏùº Îïå ADÎ•º Ï∂îÍ∞ÄÎ°ú Îçî Ï§ÑÏù¥Îäî Ïñë (dB, ÏùåÏàò)


# =======================================================
# Ïú†Ìã∏ Ìï®ÏàòÎì§
# =======================================================

def audiosegment_to_lufs(audio: AudioSegment) -> float:
    """
    pydub.AudioSegment -> integrated LUFS (EBU R128) Ï∏°Ï†ï
    """
    samples = np.array(audio.get_array_of_samples()).astype(np.float32)

    # Ïä§ÌÖåÎ†àÏò§Ïùº Í≤ΩÏö∞ Î™®ÎÖ∏Î°ú Î≥ÄÌôò
    if audio.channels > 1:
        samples = samples.reshape((-1, audio.channels))
        samples = samples.mean(axis=1)

    # 16ÎπÑÌä∏ Ïò§ÎîîÏò§Ïùò ÏµúÎåÄ ÌîºÌÅ¨ Í∞í
    peak = float(1 << (8 * audio.sample_width - 1))
    samples = samples / peak

    try:
        meter = pyln.Meter(audio.frame_rate)
        loudness = meter.integrated_loudness(samples)
        return loudness
    except Exception:
        # Ï∏°Ï†ï Ïã§Ìå® Ïãú Í∏∞Î≥∏Í∞í
        return -30.0


def parse_time_str(t: str) -> float:
    """
    "M:SS.s" ÎòêÎäî "H:MM:SS.s" ÌòïÌÉúÏùò Î¨∏ÏûêÏó¥ ÏãúÍ∞ÑÏùÑ Ï¥à(float)Î°ú Î≥ÄÌôò
    """
    parts = t.split(":")
    parts = [float(p) for p in parts]
    if len(parts) == 2:
        m, s = parts
        return m * 60 + s
    elif len(parts) == 3:
        h, m, s = parts
        return h * 3600 + m * 60 + s
    else:
        return float(t)


def sec_to_timestamp(sec: float) -> str:
    """
    Ï¥à(float) -> "M:SS.ss" Î¨∏ÏûêÏó¥
    """
    m = int(sec // 60)
    s = sec % 60
    return f"{m}:{s:05.2f}"


def change_speed(audio: AudioSegment, speed: float) -> AudioSegment:
    """
    pydubÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïò§ÎîîÏò§ ÏÜçÎèÑ Ï°∞Ï†à (ÌîºÏπò Ìï®Íªò Î≥ÄÌï®)
    """
    if speed == 1.0:
        return audio

    new_frame_rate = int(audio.frame_rate * speed)
    audio = audio._spawn(audio.raw_data, overrides={"frame_rate": new_frame_rate})
    audio = audio.set_frame_rate(16000)  # Coqui TTS Ï∂úÎ†•ÏùÄ 16kHzÎ°ú ÎßûÏ∂îÍ∏∞
    return audio


def load_mono_16k(path: str):
    """
    wavÎ•º mono / 16kÎ°ú ÎßûÏ∂∞ÏÑú torch.Tensor([T])Î°ú Î∞òÌôò (VADÏö©)
    """
    wav, sr = torchaudio.load(path)  # [C, T]
    if wav.shape[0] > 1:
        wav = wav.mean(dim=0, keepdim=True)
    if sr != 16000:
        wav = torchaudio.functional.resample(wav, sr, 16000)
        sr = 16000
    return wav.squeeze(0), sr


def get_overlap_intervals_for_ad(ad_start_sec, ad_end_sec, speech_ts, sr):
    """
    AD Íµ¨Í∞Ñ [ad_start_sec, ad_end_sec]ÏôÄ VAD speech_ts ÏÇ¨Ïù¥Ïùò
    Í≤πÏπòÎäî Íµ¨Í∞ÑÎì§ÏùÑ (Ï¥à Îã®ÏúÑ)Î°ú Í≥ÑÏÇ∞ÌïòÏó¨ Î¶¨Ïä§Ìä∏Î°ú Î∞òÌôò.
    Î∞òÌôò ÌòïÏãù: [(ov_start_sec, ov_end_sec), ...]
    """
    overlaps = []

    for seg in speech_ts:
        sp_start = seg["start"] / sr
        sp_end = seg["end"] / sr

        # Íµ¨Í∞Ñ Í≤πÏπ® Ïó¨Î∂Ä
        ov_start = max(ad_start_sec, sp_start)
        ov_end = min(ad_end_sec, sp_end)

        if ov_start < ov_end:
            overlaps.append((ov_start, ov_end))

    if not overlaps:
        return []

    # ÏãúÏûë ÏãúÍ∞Ñ Í∏∞Ï§Ä Ï†ïÎ†¨ Î∞è Î≥ëÌï©
    overlaps.sort(key=lambda x: x[0])
    merged = [overlaps[0]]

    for cur_start, cur_end in overlaps[1:]:
        last_start, last_end = merged[-1]
        if cur_start <= last_end:  # Í≤πÏπòÍ±∞ÎÇò Ïù¥Ïñ¥ÏßÄÎäî Í≤ΩÏö∞ Î≥ëÌï©
            merged[-1] = (last_start, max(last_end, cur_end))
        else:
            merged.append((cur_start, cur_end))

    return merged


# =======================================================
# 1. JSONÏóêÏÑú AD ÌååÏã± (ÏãúÍ∞Ñ Ï†ïÎ≥¥Î•º Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©)
# =======================================================

with open(JSON_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

ad_list = data["audio_descriptions"]

records = []
for item in ad_list:
    text = item["description"]
    s = parse_time_str(item["start_time"])
    e = parse_time_str(item["end_time"])

    if "duration_sec" in item:
        dur = float(item["duration_sec"])
    else:
        dur = max(0.0, e - s)

    records.append({
        "description": text,
        "start_time": s,   # Ï¥à
        "end_time": e,     # Ï¥à
        "duration_sec": dur,
    })

ad_df = pd.DataFrame(records)

# 2.5Ï¥à ÎØ∏Îßå ADÎäî ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÍ∏∞
ad_df = ad_df[ad_df["duration_sec"] >= MIN_DESC_LEN]

# ÏãúÏûë ÏãúÍ∞Ñ Í∏∞Ï§Ä Ï†ïÎ†¨
ad_df = ad_df.sort_values(by="start_time").reset_index(drop=True)

print(f"1. AD ÌååÏã± ÏôÑÎ£å. {MIN_DESC_LEN}Ï¥à Ïù¥ÏÉÅ AD Í∞úÏàò: {len(ad_df)}")

# üìå AD TIMELINE Ï∂úÎ†•
timeline_records = []
for i, row in ad_df.iterrows():
    start_sec = float(row["start_time"])
    end_sec = float(row["end_time"])
    desc = row["description"]
    timeline_records.append({
        "index": i,
        "start_time_sec": f"{start_sec:.3f}",
        "end_time_sec": f"{end_sec:.3f}",
        "start_time": sec_to_timestamp(start_sec),
        "end_time": sec_to_timestamp(end_sec),
        "description": desc,
    })
timeline_df = pd.DataFrame(timeline_records)

print("\n================ AD TIMELINE (from JSON) ================")
print(timeline_df.to_string(index=False))
print("=========================================================\n")


# =======================================================
# 2. Silero VADÎ°ú Speech Íµ¨Í∞Ñ Í≤ÄÏ∂ú (ÎåÄÏÇ¨ LUFS & Í≤πÏπ® ÌåêÎã®Ïö©)
# =======================================================

wav, sr = load_mono_16k(audio_path)
total_dur = len(wav) / sr

silero_model, silero_utils = torch.hub.load(
    repo_or_dir="snakers4/silero-vad",
    model="silero_vad",
    trust_repo=True,
)

(get_speech_timestamps,
 save_audio,
 read_audio,
 VADIterator,
 collect_chunks) = silero_utils

wav_16k = wav.float().cpu()

speech_ts = get_speech_timestamps(
    wav_16k,
    silero_model,
    sampling_rate=sr,
)

print(f"2. VAD Î∂ÑÏÑù ÏôÑÎ£å.")
print(f"  ÏõêÎ≥∏ Ïò§ÎîîÏò§ Í∏∏Ïù¥: {total_dur:.2f}Ï¥à")
print(f"  Í≤ÄÏ∂úÎêú speech Íµ¨Í∞Ñ Ïàò: {len(speech_ts)}")
print("-" * 50)


# =======================================================
# 3. Coqui XTTS v2 ÏÑ§Ï†ï Î∞è TTS ÏÉùÏÑ±
# =======================================================

TTS_MODEL_NAME = "tts_models/multilingual/multi-dataset/xtts_v2"
try:
    tts_model = TTS(TTS_MODEL_NAME, gpu=True)
    print("  ‚úÖ TTS Î™®Îç∏ GPU Î™®ÎìúÎ°ú Î°úÎìú ÏôÑÎ£å.")
except Exception as e:
    print(f"  ‚ö†Ô∏è GPU Î°úÎìú Ïã§Ìå®({e}), CPUÎ°ú Îã§Ïãú ÏãúÎèÑ.")
    tts_model = TTS(TTS_MODEL_NAME, gpu=False)

# ÎßêÌà¨/Ïñ∏Ïñ¥ ÌîÑÎ°úÌïÑ (ÌïÑÏöîÏóê Îî∞Îùº ACTIVE_PROFILEÎßå Î∞îÍøîÏÑú ÏÇ¨Ïö©)
PROFILES = {
    "kor_male_calm": {
        "language": "ko",
        "speaker_wav": "/content/drive/MyDrive/test_scene/TTS/cut_5min_kor_male.mp3",
        "speed": 1.05,
    },
    "kor_female_calm": {
        "language": "ko",
        "speaker_wav": "/content/drive/MyDrive/test_scene/TTS/cut_5min_kor_female.mp3",
        "speed": 1.05,
    },
    "eng_male_calm": {
        "language": "en",
        "speaker_wav": "/content/drive/MyDrive/test_scene/TTS/cut_5min_eng_male.mp3",
        "speed": 1.05,
    },
    "eng_female_calm": {
        "language": "en",
        "speaker_wav": "/content/drive/MyDrive/test_scene/TTS/cut_5min_eng_female.mp3",
        "speed": 1.05,
    }
}

ACTIVE_PROFILE = "eng_female_calm"  # üü¢ Ïó¨Í∏∞Îßå Î∞îÍæ∏Î©¥ ÎÇ®/Ïó¨ / Ìïú/ÏòÅ Î≥ÄÍ≤Ω Í∞ÄÎä•


def tts_synthesize(text: str, out_wav: str, profile_name: str = None):
    if profile_name is None:
        profile_name = ACTIVE_PROFILE

    cfg = PROFILES[profile_name]
    lang = cfg["language"]
    speaker_wav = cfg["speaker_wav"]
    speed = cfg["speed"]

    out_wav = Path(out_wav)
    tmp_wav = out_wav.with_suffix(".tmp.wav")

    # 1) CoquiÎ°ú TTS ÏÉùÏÑ±
    tts_model.tts_to_file(
        text=text,
        file_path=str(tmp_wav),
        speaker_wav=speaker_wav,
        language=lang,
    )

    # 2) pydubÏúºÎ°ú Î°úÎìú & Î™®ÎÖ∏/16k & ÏÜçÎèÑ Ï°∞Ï†à
    audio = AudioSegment.from_wav(str(tmp_wav))
    audio = audio.set_channels(1).set_frame_rate(16000)
    audio = change_speed(audio, speed)
    audio.export(str(out_wav), format="wav")

    if tmp_wav.exists():
        tmp_wav.unlink()


tts_raw_paths = []
for i, row in ad_df.iterrows():
    text = str(row["description"])
    raw_wav = TTS_DIR / f"ad_{i:03d}.wav"
    tts_synthesize(text, str(raw_wav))
    tts_raw_paths.append(str(raw_wav))

ad_df["tts_path"] = tts_raw_paths
print(f"3. TTS ÏÉùÏÑ± ÏôÑÎ£å. ÏÉùÏÑ± ÌååÏùº Í∞úÏàò: {len(tts_raw_paths)}")
print("-" * 50)


# =======================================================
# 4. Î∞∞Í≤Ω Î∞úÌôî LUFS Í∏∞Ï§Ä Í≥ÑÏÇ∞
# =======================================================

base_audio = AudioSegment.from_wav(audio_path)
total_ms = len(base_audio)

DIALOGUE_TRACK = AudioSegment.silent(duration=total_ms)

# Silero VAD Í≤∞Í≥ºÎ•º ÏÇ¨Ïö©Ìï¥ Dialogue Ìä∏Îûô Íµ¨ÏÑ±
for seg in speech_ts:
    # seg['start'], seg['end']Îäî ÏÉòÌîå Ïù∏Îç±Ïä§
    start_ms = int(seg['start'] * 1000 / sr)
    end_ms = int(seg['end'] * 1000 / sr)

    if start_ms < 0:
        start_ms = 0
    if end_ms > total_ms:
        end_ms = total_ms
    if start_ms >= end_ms:
        continue

    speech_seg = base_audio[start_ms:end_ms]
    DIALOGUE_TRACK = DIALOGUE_TRACK.overlay(speech_seg, position=start_ms)

try:
    if len(DIALOGUE_TRACK) < 1000:
        raise ValueError("Dialogue track too short for LUFS measurement.")

    dialogue_lufs = audiosegment_to_lufs(DIALOGUE_TRACK)
    print(f"  Î∞∞Í≤Ω Î∞úÌôî (Dialogue) ÌÜµÌï© LUFS: {dialogue_lufs:.2f} LUFS")
except Exception as e:
    dialogue_lufs = -23.0  # Í∏∞Î≥∏Í∞í
    print(f"  ‚ö†Ô∏è Î∞∞Í≤Ω Î∞úÌôî LUFS Ï∏°Ï†ï Ïã§Ìå®. Í∏∞Î≥∏Í∞í {dialogue_lufs:.2f} LUFS ÏÇ¨Ïö©: {e}")

TARGET_AD_LUFS = dialogue_lufs - DIALOGUE_LUFS_MARGIN
print(f"  üéØ AD Î™©Ìëú LUFS (Î∞∞Í≤Ω Î∞úÌôî - {DIALOGUE_LUFS_MARGIN} LUFS): {TARGET_AD_LUFS:.2f} LUFS")
print("-" * 50)


# =======================================================
# 5. Ducking ÏóÜÏù¥ AD Ïò§Î≤ÑÎ†àÏù¥ + Í≤πÏπòÎäî Î∂ÄÎ∂ÑÎßå Î∂ÄÎ∂Ñ Í∞êÏáÑ
# =======================================================

# 5-1. Ducking Ï†ÅÏö© ÏóÜÏù¥ Î∞∞Í≤Ω Ïò§ÎîîÏò§ Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©
base_audio_ducked = base_audio
print("  ‚úÖ Ducking ÏóÜÏù¥ ÏõêÎ≥∏ Î∞∞Í≤Ω Ïò§ÎîîÏò§ ÏÇ¨Ïö©.")

# 5-2. AD TTS Ìä∏Îûô ÏÉùÏÑ±
tts_track = AudioSegment.silent(duration=total_ms)
used_cnt = 0

for i, row in ad_df.iterrows():
    ad_start_sec = float(row["start_time"])        # JSONÏóêÏÑú Ïò® ÏãúÏûë ÏãúÍ∞Ñ (Ï¥à)
    tts_path = row["tts_path"]

    # ‚úÖ ADÎäî Î¨¥Ï°∞Í±¥ start_timeÏóêÎßå ÎßûÏ∂òÎã§ (end_timeÏùÄ Ï∞∏Í≥†Îßå)
    slot_start_ms = int(ad_start_sec * 1000)

    # TTS Í≤∞Í≥º Ï†ÑÏ≤¥ Í∏∏Ïù¥ ÏÇ¨Ïö© (ÏûêÎ•¥ÏßÄ ÏïäÏùå)
    ad_audio = AudioSegment.from_wav(tts_path)
    ad_duration_ms = len(ad_audio)
    ad_end_sec = ad_start_sec + (ad_duration_ms / 1000.0)

    # (1) AD Ï†ÑÏ≤¥Î•º "ÎåÄÏÇ¨ LUFS - margin" Ï†ïÎèÑÎ°ú ÎßûÏ∂§ (Í∏∞Î≥∏ Î≥ºÎ•®)
    ad_lufs = audiosegment_to_lufs(ad_audio)
    gain_db = TARGET_AD_LUFS - ad_lufs
    ad_audio = ad_audio.apply_gain(gain_db)

    # (2) Ïù¥ AD Ï†ÑÏ≤¥ Íµ¨Í∞ÑÏóê ÎåÄÌï¥, ÎåÄÏÇ¨ÏôÄ Í≤πÏπòÎäî Íµ¨Í∞Ñ(Ï¥à Îã®ÏúÑ) Î¶¨Ïä§Ìä∏ Íµ¨ÌïòÍ∏∞
    overlap_intervals_sec = get_overlap_intervals_for_ad(
        ad_start_sec, ad_end_sec, speech_ts, sr
    )

    # Í≤πÏπòÎäî Í≤å ÌïòÎÇòÎèÑ ÏóÜÏúºÎ©¥, AD Ï†ÑÏ≤¥Î•º Í∑∏ÎåÄÎ°ú overlay
    if not overlap_intervals_sec:
        tts_track = tts_track.overlay(ad_audio, position=slot_start_ms)
        used_cnt += 1
        continue

    # (3) Í≤πÏπòÎäî Íµ¨Í∞ÑÎì§ÏùÑ ms Îã®ÏúÑ (AD ÎÇ¥Î∂Ä ÏÉÅÎåÄ ÏúÑÏπò)Î°ú Î≥ÄÌôò
    # AD ÎÇ¥Î∂Ä Í∏∞Ï§Ä: 0ms ~ ad_duration_ms
    overlap_intervals_ms = []
    for ov_start_sec, ov_end_sec in overlap_intervals_sec:
        # global(sec) -> AD ÎÇ¥Î∂Ä ÏÉÅÎåÄ ms
        rel_start_ms = int((ov_start_sec - ad_start_sec) * 1000)
        rel_end_ms = int((ov_end_sec - ad_start_sec) * 1000)

        # ÌÅ¥Î¶¨Ìïë
        rel_start_ms = max(0, min(rel_start_ms, ad_duration_ms))
        rel_end_ms = max(0, min(rel_end_ms, ad_duration_ms))

        if rel_start_ms < rel_end_ms:
            overlap_intervals_ms.append((rel_start_ms, rel_end_ms))

    # (4) AD ÎÇ¥Î∂ÄÏóêÏÑú "Í≤πÏπòÏßÄ ÏïäÎäî(non-overlap) Íµ¨Í∞Ñ"Í≥º "Í≤πÏπòÎäî(overlap) Íµ¨Í∞Ñ"ÏùÑ ÎÇòÎàî
    #     - non-overlap: Í∏∞Î≥∏ Î≥ºÎ•® Ïú†ÏßÄ
    #     - overlap: AD_OVERLAP_EXTRA_ATTENUATION_DBÎßåÌÅº Ï∂îÍ∞Ä Í∞êÏáÑ
    overlap_intervals_ms.sort(key=lambda x: x[0])

    non_overlap_intervals_ms = []
    cur = 0
    for ov_start_ms, ov_end_ms in overlap_intervals_ms:
        if cur < ov_start_ms:
            non_overlap_intervals_ms.append((cur, ov_start_ms))
        cur = max(cur, ov_end_ms)
    if cur < ad_duration_ms:
        non_overlap_intervals_ms.append((cur, ad_duration_ms))

    # (5) AD Ï†ÑÏ≤¥ Í∏∏Ïù¥Ïùò "Ï°∞Ìï©Îêú" ÏÉà AudioSegment ÎßåÎì§Í∏∞
    #     - non-overlap Íµ¨Í∞Ñ: ÏõêÎûò ad_audio Í∑∏ÎåÄÎ°ú
    #     - overlap Íµ¨Í∞Ñ: ad_audioÏóêÏÑú ÏûòÎùºÏÑú Ï∂îÍ∞Ä Í∞êÏáÑ ÌõÑ Î∂ôÏù¥Í∏∞
    adjusted_ad = AudioSegment.silent(duration=ad_duration_ms)

    # 5-1) non-overlap Íµ¨Í∞Ñ Î∂ôÏù¥Í∏∞ (Í∏∞Î≥∏ Î≥ºÎ•®)
    for ns, ne in non_overlap_intervals_ms:
        if ne <= ns:
            continue
        seg = ad_audio[ns:ne]
        adjusted_ad = adjusted_ad.overlay(seg, position=ns)

    # 5-2) overlap Íµ¨Í∞Ñ Î∂ôÏù¥Í∏∞ (Î∂ÄÎ∂Ñ Í∞êÏáÑ)
    for os_ms, oe_ms in overlap_intervals_ms:
        if oe_ms <= os_ms:
            continue
        seg = ad_audio[os_ms:oe_ms]
        seg = seg.apply_gain(AD_OVERLAP_EXTRA_ATTENUATION_DB)
        adjusted_ad = adjusted_ad.overlay(seg, position=os_ms)

    # (6) ÏµúÏ¢Ö ADÎ•º start_time ÏúÑÏπòÏóê overlay
    tts_track = tts_track.overlay(adjusted_ad, position=slot_start_ms)
    used_cnt += 1

print(f"  Ïã§Ï†úÎ°ú ÏÇ¨Ïö©Îêú AD Í∞úÏàò: {used_cnt}")
print("-" * 50)


# =======================================================
# 6. ÏµúÏ¢Ö ÎØπÏã± Î∞è ÌååÏùº Ï†ÄÏû•
# =======================================================

mixed_audio = base_audio_ducked.overlay(tts_track)
mixed_audio.export(final_audio_path, format="wav")
print(f"6. ÏµúÏ¢Ö AD Ïò§ÎîîÏò§ Ï†ÄÏû• ÏôÑÎ£å: {final_audio_path}")


# =======================================================
# 7. ffmpeg-pythonÏúºÎ°ú mp4Ïóê ÏÉà Ïò§ÎîîÏò§ ÏûÖÌûàÍ∏∞
# =======================================================

video_in = ffmpeg.input(input_video)
audio_in = ffmpeg.input(final_audio_path)

out = ffmpeg.output(
    video_in.video,
    audio_in.audio,
    output_video,
    vcodec="copy",
    acodec="aac",
).overwrite_output()

out.run()
print(f"7. ÏµúÏ¢Ö AD ÎπÑÎîîÏò§ ÏÉùÏÑ± ÏôÑÎ£å: {output_video}")
```
